Importação de bibliotaca para carregar e mostrar a imagem

```
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

Importação da imagem

```
# Caminho da imagem
Caminho_Imagem = '/content/Imagens_Testar/lenna.jpg'

# Usando a CV2 apenas poara carregar a imagem e transformar em RGB
image = cv2.cvtColor(cv2.imread(Caminho_Imagem), cv2.COLOR_BGR2RGB)

print("Imagem Normal")
plt.imshow(image)
plt.axis('off')
plt.show()
```

Redução de tamanho
